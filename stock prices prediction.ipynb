```python
#Training and splitting the dataset i.e. last (1 year into validation)
train=avgdata[:1670]
test=avgdata[1670:]
```

```python
test.shape
```

```python
#Predictions using Moving Average
predictions=[]
for i in range(0,365):
    x=train['Close'][1305+i:].sum()+sum(predictions)
    y=x/365
    predictions.append(y)
```
